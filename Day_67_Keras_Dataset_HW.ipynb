```python
#導入影像列印模組
import matplotlib.pyplot as plt

def plot_images_labels_prediction(images,labels,num=10):
    fig =plt.figure(figsize=(10,10))

    for i in range(num):
        ax = fig.add_subplot(2, int(num/2), i+1, xticks = [], yticks = [])
        ax.imshow(images[i], cmap='binary')
        ax.set_title(labels[i],fontsize=10)

    plt.subplots_adjust(left=0.125, bottom=0.0
                        , right=0.9, top=0.3) 
    plt.show()
plot_images_labels_prediction(x_img_train,y_label_train,10)
```
