1. LASSO 回歸可以被用來作為 Feature selection 的工具，請了解 LASSO 模型為什麼可用來作 Feature selection


LASSO為回歸加上L1正則，而L1正則化會把系數w的l1範數作為懲罰項加到損失函數上，所以當損失函數求導數的時候，L1得到的值會很少，這個特性使得L1正則化成為一種很好的特征選擇方法。


2. 當自變數 (X) 存在高度共線性時，Ridge Regression 可以處理這樣的問題嗎?

L2懲罰參數的值可以透過「tuninig parameter, 𝜆」來控制。當𝜆→0，L2懲罰參數就跟OLS回歸一樣，目標函式只有最小化SSE；而當𝜆→∞時，懲罰效果最大，迫使所有係數都趨近於0。
